### Load Data

```python
# Load CDOT Data for 2021 - 2024

years = [2021, 2022, 2023, 2024]
cdot_base_path = "/Users/haley.northrup/Desktop/PeopleCenteredCOS/cdot_data/"

loaded_dfs = []
for year in years:
    filename = f"CDOTRM_CD_Crash_Listing_-_{year}.xlsx"

    full_path = cdot_base_path + filename
    print(full_path)

    temp_cdot_pdf = pd.read_excel(cdot_base_path + filename)
    loaded_dfs.append(temp_cdot_pdf)

# combine sheets 
cdot_pdf = pd.concat(loaded_dfs)

print(cdot_pdf['Crash Date'].nunique())
```

```python
# limit data for quicker iteration *** 
columns_of_interest = ['CUID', 'Crash Date', 'Crash Time', 'City', 'County', 'Latitude', 'Longitude', 'Location 1', 'Link', 'Location 2', 'Number Killed']

#cdot_pdf_small = cdot_pdf.head(10000)[columns_of_interest]
cdot_pdf_slim = cdot_pdf[columns_of_interest]

display(cdot_pdf_slim)
```

### Filter and Clean Data 

```python
# Add new column for truncated Latitude and Longitude 
# Latitude and Longitude are in decimal degrees - https://en.wikipedia.org/wiki/Decimal_degrees
# To identify an "individual street, large building" use 4 decimal places 
# "individual trees, houses" - 5 decimal places 

cdot_pdf_slim.loc[:, 'Latitude_4dec'] = cdot_pdf_slim['Latitude'].round(4)
cdot_pdf_slim.loc[:, 'Longitude_4dec'] = cdot_pdf_slim['Longitude'].round(3)
cdot_pdf_slim.loc[:, 'Latitude_3dec'] = cdot_pdf_slim['Latitude'].round(3)
cdot_pdf_slim.loc[:, 'Longitude_3dec'] = cdot_pdf_slim['Longitude'].round(3)
```

```python
# Only include El Paso County 

cdot_elpaso_pdf = cdot_pdf_slim.loc[cdot_pdf_slim['County'] == 'EL PASO']

print(cdot_elpaso_pdf.shape)
display(cdot_elpaso_pdf.head(5))
```

## Analyze Data 

### Crashes by Location 

#### Explore Crash Data Locations by Location 1 + Location 2

```python
# combine location 1 and location 2
# more useful to understand the intersection vs. location 1 at location 2 being different than location 2 at location 2. 
# *************************

# determine if Location 1 is first alphabetically
cdot_elpaso_pdf.loc[:, 'loc1_first'] = cdot_elpaso_pdf.loc[:, 'Location 1'] > cdot_elpaso_pdf.loc[:, 'Location 2'] 

# combine Location 1 and Location 2 alphabetically 
# location 1 is first
cdot_elpaso_pdf.loc[cdot_elpaso_pdf['loc1_first'], 'combine_alph_loc1_loc2'] = \
        cdot_elpaso_pdf.loc[cdot_elpaso_pdf['loc1_first'], 'Location 1'] + \
        '    ' + \
        cdot_elpaso_pdf.loc[cdot_elpaso_pdf['loc1_first'], 'Location 2']
# location 2 is second 
cdot_elpaso_pdf.loc[~cdot_elpaso_pdf['loc1_first'], 'combine_alph_loc1_loc2'] = \
        cdot_elpaso_pdf.loc[~cdot_elpaso_pdf['loc1_first'], 'Location 2'] + \
        '    ' + \
        cdot_elpaso_pdf.loc[~cdot_elpaso_pdf['loc1_first'], 'Location 1']

# check result
display(cdot_elpaso_pdf[['Location 1', 'Location 2', 'loc1_first', 'combine_alph_loc1_loc2']].head(5))
```

```python
print(cdot_elpaso_pdf.groupby(['combine_alph_loc1_loc2']).count()['CUID'].sort_values(ascending=False).shape)

display(cdot_elpaso_pdf.groupby(['combine_alph_loc1_loc2']).count()['CUID'].sort_values(ascending=False).head(30))
```
