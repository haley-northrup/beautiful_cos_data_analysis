```python
# Rd_Number

import numpy as np

# set Rd_Number to none since I'm not sure how to do these mappings
cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['City Street', 'County Road']), 'Rd_Number'] = np.nan
display(cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['City Street', 'County Road'])])

# Rd_Number is the RTE (with 3 digits) and the SEC
cdot_old_converted_pdf['RTE_3dig'] = cdot_old_converted_pdf['RTE'].apply(lambda x: f"{x:03}")
cdot_old_converted_pdf['RTE_3dig_SEC'] = cdot_old_converted_pdf['RTE_3dig'] + cdot_old_converted_pdf['SEC']
# display(cdot_old_converted_pdf[['RTE', 'RTE_3dig','RTE_3dig_SEC']])

cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['Interstate Highway', 'State Highway', 'Frontage Road']), 'Rd_Number'] = \
    cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['Interstate Highway', 'State Highway', 'Frontage Road']), 'RTE_3dig_SEC'] 

display(cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['Interstate Highway', 'State Highway', 'Frontage Road'])])
```

```python
# Rd_Section 
import re

def convert_location1_to_rd_number(LOC01):
    drop_numbers = re.sub(r'\\d+\\s', '', str(LOC01))
    drop_single_letters = re.sub('\\s[a-z]\\s', '', drop_numbers)
    split_list = drop_single_letters.split()
    if len(split_list) == 0:
        return np.nan
    else:
        get_first_word = split_list[0]
        first_word_first_five_letters_uppercase = get_first_word[0:5].upper()
        return first_word_first_five_letters_uppercase
    
# set Rd_Section to part of location 1 for City Street and County Road
cdot_old_converted_pdf['Location 1'] = cdot_old_pdf['LOC_01']
cdot_old_converted_pdf['Location 1 converted to Rd_Section'] = cdot_old_converted_pdf['Location 1'].apply(convert_location1_to_rd_number)
cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['City Street', 'County Road']), 'Rd_Section'] = \
    cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['City Street', 'County Road']), 'Location 1 converted to Rd_Section']

# Rd_Section is MP for Highways and Frontage roads
cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['Interstate Highway', 'State Highway', 'Frontage Road']), 'Rd_Section'] = \
    cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['Interstate Highway', 'State Highway', 'Frontage Road']), 'MP'] 

display(cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['City Street', 'County Road'])])
display(cdot_old_converted_pdf.loc[cdot_old_converted_pdf['System Code'].isin(['Interstate Highway', 'State Highway', 'Frontage Road'])])
```

```python
def convert_4digit_time_to_timestr(time_4digit):
    add_colons_and_seconds = time_4digit[0:2] + ':' + time_4digit[2:4] + ':00' 
    return add_colons_and_seconds
    
cdot_old_converted_pdf['TIME'] = cdot_old_pdf['TIME']
cdot_old_converted_pdf['TIME_int_str'] = cdot_old_converted_pdf['TIME'].fillna(-1).astype(int).astype(str)
cdot_old_converted_pdf['Crash Time'] = cdot_old_converted_pdf['TIME_int_str'].apply(lambda x: f"{x:04}") \
    .apply(convert_4digit_time_to_timestr)
display(cdot_old_converted_pdf['Crash Time'])
```
