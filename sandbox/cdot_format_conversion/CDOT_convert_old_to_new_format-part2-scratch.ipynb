### Identify Motorists and Non-Motorists 
#### Populate Type Columns 

```python
# Identify which VEHICLE_X columns need to be converted to NM columns in new format 

# When the accident type is Pedestrian the vast majority of the time the "VEHICLE_X" column that says "OTHER - SEE REPORT" is referring to a pedestrian 
# there are a handful of accidents that have multiple "VEHICLE_X" columns that say "OTHER - SEE REPORT" so it is ambiguous which X is the pedestrian 

cdot_old_pdf['vehicle_1_is_nm'] = False
cdot_old_pdf['vehicle_2_is_nm'] = False
cdot_old_pdf['vehicle_3_is_nm'] = False

#cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_1'] == np.nan, 'vehicle_1_is_nm'] = np.nan 
#cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_2'] == np.nan, 'vehicle_2_is_nm'] = np.nan 
#cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_3'] == np.nan, 'vehicle_3_is_nm'] = np.nan 

cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_1'] == 'OTHER - SEE REPORT'), 'vehicle_1_is_nm'] = True 
cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_2'] == 'OTHER - SEE REPORT'), 'vehicle_2_is_nm'] = True
cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_3'] == 'OTHER - SEE REPORT'), 'vehicle_3_is_nm'] = True

cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_1'] == 'OTHER - SEE REPORT'), 'TU-1 NM Type'] = 'Pedestrian'
cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_2'] == 'OTHER - SEE REPORT'), 'TU-2 NM Type'] = 'Pedestrian'
cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_3'] == 'OTHER - SEE REPORT'), 'TU-3 NM Type'] = 'Pedestrian'

display(cdot_old_pdf.loc[cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN', 
['ACCTYPE', 
 'VEHICLE_1', 'vehicle_1_is_nm', 'TU-1 NM Type', 
 'VEHICLE_2', 'vehicle_2_is_nm', 'TU-2 NM Type', 
 'VEHICLE_3', 'vehicle_3_is_nm', 'TU-3 NM Type']])

# Old format has Bicycle as a type of vehicle, so filter for VEHICLE_X == BICYCLE or MOTORIZED BICYCLE, then slot those in as non-motorists 
# accident type isn't always bicycle when a bicycle is involved 

cdot_old_pdf.loc[(cdot_old_pdf['VEHICLE_1'] == 'BICYCLE') | (cdot_old_pdf['VEHICLE_1'] == 'MOTORIZED BICYCLE'), 'vehicle_1_is_nm'] = True
cdot_old_pdf.loc[(cdot_old_pdf['VEHICLE_2'] == 'BICYCLE') | (cdot_old_pdf['VEHICLE_2'] == 'MOTORIZED BICYCLE'), 'vehicle_2_is_nm'] = True
cdot_old_pdf.loc[(cdot_old_pdf['VEHICLE_3'] == 'BICYCLE') | (cdot_old_pdf['VEHICLE_3'] == 'MOTORIZED BICYCLE'), 'vehicle_3_is_nm'] = True

cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_1'] == 'BICYCLE', 'TU-1 NM Type'] = 'Bicyclist'
cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_2'] == 'BICYCLE', 'TU-2 NM Type'] = 'Bicyclist'
cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_3'] == 'BICYCLE', 'TU-3 NM Type'] = 'Bicyclist'
cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_1'] == 'MOTORIZED BICYCLE', 'TU-1 NM Type'] = 'Other Bicyclist/Cyclist'
cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_2'] == 'MOTORIZED BICYCLE', 'TU-2 NM Type'] = 'Other Bicyclist/Cyclist'
cdot_old_pdf.loc[cdot_old_pdf['VEHICLE_3'] == 'MOTORIZED BICYCLE', 'TU-3 NM Type'] = 'Other Bicyclist/Cyclist'

display(cdot_old_pdf.loc[cdot_old_pdf['ACCTYPE'] == 'BICYCLE', 
['ACCTYPE', 'VEHICLE_1', 'vehicle_1_is_nm', 'TU-1 NM Type', 
 'VEHICLE_2', 'vehicle_2_is_nm', 'TU-2 NM Type',
 'VEHICLE_3', 'vehicle_3_is_nm', 'TU-3 NM Type']])
```

```python
# Motorist Type Mapping

# Convert VEHICLE_X (old column) to TU-X Type (new column) - when VEHICLE_X is a motorist 

# And segment the Non-motorists out of the old format and note which vehicle number is the non-motorist if any 
old_column = 'VEHICLE_1'
new_column = 'TU-1 Type'
new_column2 = 'TU-1 NM Type'

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()
print(f"New format {new_column2}", cdot_new_pdf[new_column2].unique())

# don't map bicycle or motorized bicycle since they will always be mapped to non-motorists 
map_vehicle_to_tu_type_x = {'HIT & RUN - UNKNOWN': 'Unknown', 
                 'UNKNOWN': 'Unknown',
                 'PICKUP TRUCK/UTILITY VAN': 'Pickup Truck/Utility Van', 
                 'PASSENGER CAR/VAN': 'Passenger Car/Passenger Van', 
                 'SUV': 'SUV', 
                 'MOTORCYCLE': 'Motorcycle', 
                 'PASSENGER CAR/VAN W/TRAILER': 'Passenger Car/Passenger Van with Trailer - RET',
                 'SCHOOL BUS < 15 PEOPLE': 'School Bus (all school buses)',
                 'FARM EQUIPMENT': 'Farm Equipment',
                 'MOTOR HOME': 'Motor Home',
                 'TRUCK GVW > 10K/BUSSES > 15 PEOPLE': 'old__truck_gvw_gr_10k_busses_gr_15_people',
                 'SUV W/TRAILER': 'old__SUV_w_trailer',
                 'PICKUP TRUCK/UTILITY VAN W/TRAILER': 'old__pickup_truck_utility_van_w_trailer',
                 'NON-SCHOOL BUS < 15 PEOPLE': '',
                 'OTHER - SEE REPORT': 'Other Vehicle Type (Describe in Narrative)',
                }

# In the new data format, if the accident is between a motorist and a non-motorist
# there is no re-use of the 1, 2 numbers 
# TU-1 Type = Null 
# TU-2 Type = SUV
# TU-1 NM = Pedestrian

cdot_old_pdf.loc[cdot_old_pdf['vehicle_1_is_nm'] == False, 'TU-1 Type'] = cdot_old_pdf['VEHICLE_1'].map(map_vehicle_to_tu_type_x)
cdot_old_pdf.loc[cdot_old_pdf['vehicle_2_is_nm'] == False, 'TU-2 Type'] = cdot_old_pdf['VEHICLE_2'].map(map_vehicle_to_tu_type_x) 
cdot_old_pdf.loc[cdot_old_pdf['vehicle_3_is_nm'] == False, 'TU-3 Type'] = cdot_old_pdf['VEHICLE_3'].map(map_vehicle_to_tu_type_x) 

display(cdot_old_pdf[['ACCTYPE', 
                      'VEHICLE_1', 'vehicle_1_is_nm', 'TU-1 Type', 'TU-1 NM Type', 
                      'VEHICLE_2', 'vehicle_2_is_nm', 'TU-2 Type', 'TU-2 NM Type',
                      'VEHICLE_3', 'vehicle_3_is_nm', 'TU-3 Type', 'TU-3 NM Type']])

print(f"New converted format {new_column2}", cdot_old_pdf[new_column2].unique())
print(f"New converted format {new_column}", cdot_old_pdf[new_column].unique())
```

```python
# Direction 
# ************************

# Convert DIR_X (old column) to TU-X Direction and TU-X NM Direction (new column)
# Convert enum values

# Direction
old_column = 'DIR_1'
new_column = 'TU-1 Direction'
new_column2 = 'TU-1 NM Direction'

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()
print(f"New format {new_column2}", cdot_new_pdf[new_column2].unique())

map_direction_x = {'S': 'South', 
                   'N': 'North',
                   'W': 'West',
                   'UK': 'Unknown',
                   'E': 'East',
                   'SE': 'Southeast',
                   'NW': 'Northwest',
                   'NE': 'Northeast',
                   'SW': 'Southwest',
                }

for x in ['1', '2', '3']:
    cdot_old_pdf.loc[cdot_old_pdf[f"vehicle_{x}_is_nm"] == True, f"TU-{x} NM Direction"] = cdot_old_pdf[f"DIR_{x}"].map(map_direction_x)
    cdot_old_pdf.loc[cdot_old_pdf[f"vehicle_{x}_is_nm"] == False, f"TU-{x} Direction"] = cdot_old_pdf[f"DIR_{x}"].map(map_direction_x)

print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())
print(f"Old converted format {new_column2}", cdot_old_pdf[new_column2].unique())

display(cdot_old_pdf[['DIR_1', 'vehicle_1_is_nm', 'TU-1 Direction', 'TU-1 NM Direction',
                      'DIR_2', 'vehicle_2_is_nm', 'TU-2 Direction', 'TU-2 NM Direction',
                      'DIR_3', 'vehicle_3_is_nm', 'TU-3 Direction', 'TU-3 NM Direction']])
```

```python
# Convert VEHICLE_X (old column) to TU-X Hit And Run
# Convert enum values

# Movement
old_column = 'VEHICLE_1'
new_column = 'TU-1 Hit And Run'

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()


for x in ['1', '2', '3']:
    condition_hit_and_run__true = (cdot_old_pdf[f"vehicle_{x}_is_nm"] == False) & (cdot_old_pdf[f"VEHICLE_{x}"] == "HIT & RUN - UNKNOWN")
    condition_hit_and_run__false = (cdot_old_pdf[f"vehicle_{x}_is_nm"] == False) & (cdot_old_pdf[f"VEHICLE_{x}"] != "HIT & RUN - UNKNOWN")
    cdot_old_pdf.loc[condition_hit_and_run__true, f"TU-{x} Hit And Run"] = True
    cdot_old_pdf.loc[condition_hit_and_run__false, f"TU-{x} Hit And Run"] = False
    
print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())

display(cdot_old_pdf[['VEHICLE_1', 'vehicle_1_is_nm', 'TU-1 Hit And Run', 
                      'VEHICLE_2', 'vehicle_2_is_nm', 'TU-2 Hit And Run', 
                      'VEHICLE_3', 'vehicle_3_is_nm', 'TU-3 Hit And Run']])
```

```python
# Convert LIMIT_X (old column) to TU-X Speed Limit (new column)
# Convert data type

old_column = 'LIMIT1'
new_column = 'TU-1 Speed Limit'

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()

for x in ['1', '2', '3']:
    cdot_old_pdf.loc[cdot_old_pdf[f"vehicle_{x}_is_nm"] == False, f"TU-{x} Speed Limit"] = cdot_old_pdf[f"LIMIT{x}"].replace('UK', '-1').astype(int).replace(-1, np.nan)

print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())

display(cdot_old_pdf[['VEHICLE_1', 'vehicle_1_is_nm', 'LIMIT1', 'TU-1 Speed Limit', 
                      'VEHICLE_2', 'vehicle_2_is_nm', 'LIMIT2', 'TU-2 Speed Limit', 
                      'VEHICLE_3', 'vehicle_3_is_nm', 'LIMIT3', 'TU-3 Speed Limit']])
```

```python
# Convert SPEED_X (old column) to TU-X Estimated Speed (new column)
# Convert using regular expressions - remove leading zeros, convert to int, 
# handle 'UK' --> placeholder --> nan

import re 

old_column = 'SPEED_1'
new_column = 'TU-1 Estimated Speed'

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()

for x in ['1', '2', '3']:
    cdot_old_pdf.loc[cdot_old_pdf[f"vehicle_{x}_is_nm"] == False, f"TU-{x} Estimated Speed"] = cdot_old_pdf[f"SPEED_{x}"].replace('UK', '-1').replace(np.nan, '-1').str.replace(r'^0+(?!\s*$)', '', regex=True).astype('int').replace(-1, np.nan)

print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())

display(cdot_old_pdf[['VEHICLE_1', 'vehicle_1_is_nm', 'SPEED_1', 'TU-1 Estimated Speed', 
                      'VEHICLE_2', 'vehicle_2_is_nm', 'SPEED_2', 'TU-2 Estimated Speed', 
                      'VEHICLE_3', 'vehicle_3_is_nm', 'SPEED_3', 'TU-3 Estimated Speed']])
```
