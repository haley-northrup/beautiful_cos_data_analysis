```python
# Convert FACTOR_X (old column) to 
# TU-X Human Contributing Factor and TU-X NM Human Contributing Factor (new column)
# Convert enums

old_column = 'FACTOR_1'
new_column = 'TU-1 Human Contributing Factor'
new_column2 = 'TU-1 NM Human Contributing Factor '

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()
print(f"New format {new_column2}", cdot_new_pdf[new_column2].unique())
print()


map_factor_x = {'NO APPARENT CONTRIBUTING FACTOR': 'No Apparent Contributing Factor',
                'UNKNOWN': np.nan,
                'DRIVER UNFAMILIAR WITH AREA': 'Driver Unfamiliar With Area',
                'DRIVER PREOCCUPIED':'Distracted',
                'DRIVER INEXPERIENCE': 'Driver Inexperience',
                'DRIVER FATIGUE': 'Asleep or Fatigued',
                'ASLEEP AT WHEEL': 'Asleep at the wheel',
                'ILLNESS': 'Illness',
                'DRIVER EMOTIONALLY UPSET': 'Driver Emotionally Upset',
                'DISTRACTED BY PASSENGER': 'Distracted/Other Occupant',
                'EVADING LAW ENFORCEMENT OFFICER': 'Evading Law Enforcement Officer',
                'PHYSICAL DISABILITY': 'Physical Disability',
               }

cdot_old_pdf[new_column] = cdot_old_pdf[old_column].map(map_factor_x)
print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())
```

```python
# Convert AGE_X (old column) to TU-X Age and TU-X NM Age (new column)
# Rename column

old_column = 'AGE_1'
new_column = 'TU-1 Age'
new_column2 = 'TU-1 NM Age '

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()
print(f"New format {new_column2}", cdot_new_pdf[new_column2].unique())
print()

cdot_old_pdf[new_column] = cdot_old_pdf[old_column]
print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())
```

```python
# Convert BELT_X (old column) to TU-X Safety restaint Use (new column)
 # Map values

old_column = 'BELT_1'
new_column = 'TU-1 Safety restraint Use'

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()

map_belt_x = {'Y': 'old__yes',
                'N': 'old__no',
               }

cdot_old_pdf[new_column] = cdot_old_pdf[old_column].map(map_belt_x)
print(f"Old converted format {new_column}", cdot_old_pdf[new_column].unique())
```

### Investigate Alcohol, Marijuana and Other Drugs columns

```python
# Convert DRIVER_X and DUI_X (old column) to TU-X Alcohol Suspected and TU-X NM Alcohol Suspected (new column)
# Convert enums

old_column = 'DRIVER_1'
old_column2 = 'DUI_1'
new_column = 'TU-1 Alcohol Suspected'
new_column2 = 'TU-1 NM Alcohol Suspected '
new_column3 = 'TU-1  Marijuana Suspected'
new_column4 = 'TU-1 NM Marijuana Suspected '
new_column5 = 'TU-1 Other Drugs Suspected '
new_column6 = 'TU-1 NM Other Drugs Suspected '

map_alcohol_drugs_x = {}

print(f"Old format {old_column}", cdot_old_pdf[old_column].unique())
print()
print(f"Old format {old_column2}", cdot_old_pdf[old_column2].unique())
print()
print(f"New format {new_column}", cdot_new_pdf[new_column].unique())
print()
print(f"New format {new_column2}", cdot_new_pdf[new_column2].unique())
print()
print(f"New format {new_column3}", cdot_new_pdf[new_column3].unique())
print()
print(f"New format {new_column4}", cdot_new_pdf[new_column4].unique())
print()
print(f"New format {new_column5}", cdot_new_pdf[new_column5].unique())
print()
print(f"New format {new_column6}", cdot_new_pdf[new_column6].unique())
print()
```

```python
# Mappings
# OLD: NO IMPAIRMENT SUSPECTED 
# --> NEW Alcohol Suspected - "No" 
# --> NEW Marijuana Suspected - "Marijuana Not Suspected"
# --> NEW Other Drugs Suspected - "No" 

# OLD: ALCOHOL INVOLVED 
# --> New Alcohol Suspected - "Yes" 
# --> NEW Marijuana Suspected - "Unknown"
# --> NEW Other Drugs Suspected - "Unknown" 

# OLD: ALCOHOL/DRUGS 
# --> New Alcohol Suspected - "old__alcohol_drugs" 
# --> NEW Marijuana Suspected - "old__alcohol_drugs"
# --> NEW Other Drugs Suspected - "old__alcohol_drugs"

# OLD: RX/MEDICATION/DRUGS 
# --> New Alcohol Suspected - "No" 
# --> NEW Marijuana Suspected - "old__rx_medication_drugs"
# --> NEW Other Drugs Suspected - "old__rx_medication_drugs"

cdot_old_pdf['Alcohol Suspected'] = np.nan
cdot_old_pdf['Marijuana Suspected'] = np.nan
cdot_old_pdf['Other Drugs Suspected'] = np.nan 

cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'NO IMPAIRMENT SUSPECTED', 'Alcohol Suspected'] = "No"
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'NO IMPAIRMENT SUSPECTED', 'Marijuana Suspected'] = 'Marijuana Not Suspected'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'NO IMPAIRMENT SUSPECTED', 'Other Drugs Suspected'] = 'No'

cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'ALCOHOL INVOLVED', 'Alcohol Suspected'] = 'Yes'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'ALCOHOL INVOLVED', 'Marijuana Suspected'] = 'Unknown'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'ALCOHOL INVOLVED', 'Other Drugs Suspected'] = 'Unknown'

cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'ALCOHOL/DRUGS', 'Alcohol Suspected'] = 'old__alcohol_drugs'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'ALCOHOL/DRUGS', 'Marijuana Suspected'] = 'old__alcohol_drugs'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'ALCOHOL/DRUGS', 'Other Drugs Suspected'] = 'old__alcohol_drugs'

cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'RX/MEDICATION/DRUGS', 'Alcohol Suspected'] = 'No'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'RX/MEDICATION/DRUGS', 'Marijuana Suspected'] = 'old__rx_medication_drugs'
cdot_old_pdf.loc[cdot_old_pdf['DRIVER_1'] == 'RX/MEDICATION/DRUGS', 'Other Drugs Suspected'] = 'old__rx_medication_drugs'
```
