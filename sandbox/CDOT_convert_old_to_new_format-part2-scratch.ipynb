```python
# Identify which VEHICLE_X columns need to be converted to NM columns in new format 

# When the accident type is Pedestrian the vast majority of the time the "VEHICLE_X" column that says "OTHER - SEE REPORT" is referring to a pedestrian 
# there are a handful of accidents that have multiple "VEHICLE_X" columns that say "OTHER - SEE REPORT" so it is ambiguous which X is the pedestrian 

cdot_old_pdf['vehicle_1_is_nm'] = ''
cdot_old_pdf['vehicle_2_is_nm'] = ''
cdot_old_pdf['vehicle_3_is_nm'] = ''

cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_1'] == 'OTHER - SEE REPORT'), 'vehicle_1_is_nm'] = 'Pedestrian'
cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_2'] == 'OTHER - SEE REPORT'), 'vehicle_2_is_nm'] = 'Pedestrian'
cdot_old_pdf.loc[(cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN') & (cdot_old_pdf['VEHICLE_3'] == 'OTHER - SEE REPORT'), 'vehicle_3_is_nm'] = 'Pedestrian'

display(cdot_old_pdf.loc[cdot_old_pdf['ACCTYPE'] == 'PEDESTRIAN', 
['ACCTYPE', 'VEHICLE_1', 'vehicle_1_is_nm', 'VEHICLE_2', 'vehicle_2_is_nm', 'VEHICLE_3', 'vehicle_3_is_nm']])

# Old format has Bicycle as a type of vehicle, so filter for VEHICLE_X == BICYCLE or MOTORIZED BICYCLE, then slot those in as non-motorists 
# accident type isn't always bicycle when a bicycle is involved 

cdot_old_pdf.loc[(cdot_old_pdf['VEHICLE_1'] == 'BICYCLE') | (cdot_old_pdf['VEHICLE_1'] == 'MOTORIZED BICYCLE'), 'vehicle_1_is_nm'] = 'Bicyclist'
cdot_old_pdf.loc[(cdot_old_pdf['VEHICLE_2'] == 'BICYCLE') | (cdot_old_pdf['VEHICLE_2'] == 'MOTORIZED BICYCLE'), 'vehicle_2_is_nm'] = 'Bicyclist'
cdot_old_pdf.loc[(cdot_old_pdf['VEHICLE_3'] == 'BICYCLE') | (cdot_old_pdf['VEHICLE_3'] == 'MOTORIZED BICYCLE'), 'vehicle_3_is_nm'] = 'Bicyclist'

display(cdot_old_pdf.loc[cdot_old_pdf['ACCTYPE'] == 'BICYCLE', 
['ACCTYPE', 'VEHICLE_1', 'vehicle_1_is_nm', 'VEHICLE_2', 'vehicle_2_is_nm', 'VEHICLE_3', 'vehicle_3_is_nm']])
```

```python
# In the new data format, if the accident is between a motorist and a non-motorist
# there is no re-use of the 1, 2 numbers 
# TU-1 Type = Null 
# TU-2 Type = SUV
# TU-1 NM = Pedestrian 

# If VEHICLE_X is a non-motorist, then all COL_X columns are converted to TU-X NM columns 

# set non-motorist type to be the vehicle_x_is_nm indicator column
cdot_old_pdf['TU-1 NM Type'] = cdot_old_pdf['vehicle_1_is_nm']
cdot_old_pdf['TU-2 NM Type'] = cdot_old_pdf['vehicle_2_is_nm']
cdot_old_pdf['TU-3 NM Type'] = cdot_old_pdf['vehicle_3_is_nm']

# set the motorist type (TU-X Type) to be empty when vehicle_x_is_nm is not empty
cdot_old_pdf.loc[cdot_old_pdf['vehicle_1_is_nm'] != '', 'TU-1 Type'] = ''
cdot_old_pdf.loc[cdot_old_pdf['vehicle_2_is_nm'] != '', 'TU-2 Type'] = ''
cdot_old_pdf.loc[cdot_old_pdf['vehicle_3_is_nm'] != '', 'TU-3 Type'] = ''

display(cdot_old_pdf.loc[cdot_old_pdf['ACCTYPE'] == 'BICYCLE', 
['ACCTYPE', 'VEHICLE_1', 'TU-1 Type', 'TU-1 NM Type', 'VEHICLE_2', 'TU-2 Type', 'TU-2 NM Type']])
```
